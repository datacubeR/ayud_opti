```python
from sklearn.datasets import load_iris
import numpy as np
np.random.seed(42)

A = load_iris()["data"]
A
```

```python
# PCA es que vamos a calcular los valores y vectores propios de una matriz de covarianza generada con datos centrados...


def pca_eig(A, k):
    A_center = A - A.mean(axis=0) # Broadcasting... A.mean(axis = 0) (1x4) A 150x4
    n = A_center.shape[0]
    S = 1/(n-1) * A_center.T @ A_center # 150x4

    # PCA
    # valores propios representan la varianza explicada de cada componente. 
    # Vectores propios representan las componentes principales. 

    var, C = np.linalg.eig(S)

    # Ordenamos los valores y vectores de mayor a menor...
    idx = np.argsort(-var)
    var = var[idx]/var.sum()
    C = C[:,idx] # columnas son los CP

    if 1 <= k <= A_center.shape[1]:
        return C.T[:k], var[:k]
    elif 0 < k < 1:
        idx = np.argmax(var.cumsum() > k) + 1
        return C.T[:idx], var[:idx]
```

```python
from sklearn.decomposition import PCA

for k in [2,3,0.95,0.98]:
    pca = PCA(n_components=k)
    pca.fit(A)
    C = pca.components_
    var_exp = pca.explained_variance_ratio_
    print("---------------------------------")
    print(f"PCA SKlearn: Iris, k = {k}")
    print("---------------------------------")
    print(f"C = {C}")
    print("---------------------------------")
    print(f"Var Explicada = {var_exp}")
```

```python
for k in [2,3,0.95,0.98]:
    C, var_exp = pca_eig(A, k)
    print("---------------------------------")
    print(f"Nuestra Implementación: Iris, k = {k}")
    print("---------------------------------")
    print(f"C = {C}")
    print("---------------------------------")
    print(f"Var Explicada = {var_exp}")
```
