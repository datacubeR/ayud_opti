# Pregunta 2

```python
A = np.array([
    [2,4,-2],
    [-2,-3,7],
    [3,9,-4]
])

L2 = np.array([
    [1,0,0],
    [0,1,0],
    [0,-3,1] # a la fila 3 = fila 3 - 3 veces la fila 2. 
])

L1 = np.array([
    [1,0,0],
    [1,1,0], # fila 2 = fila 2 + fila 1
    [-3/2,0,1] # fila 3 = fila 3 - 3/2 fila 1
])

# L2 * L1 * A = U

U = L2 @ L1 @ A
U
```

```python
L = np.linalg.inv(L2 @ L1)
L @ U
```

```python
L @ U
```

# Pregunta 3

```python
from sklearn.datasets import load_iris

A = load_iris()["data"]
A
```

```python
k = 2

def pca_eig(A,k):
    A_center = A - A.mean(axis = 0)
    S = 1 / (A.shape[0] - 1) * A_center.T @ A_center
    var, P = np.linalg.eig(S)

    idx = np.argsort(-var)

    var, P = var[idx]/var.sum(), P[:,idx]
    if 1 <= k <= A.shape[1]:
        return P.T[:k], var[:k]
    elif 0 < k < 1:
        idx = np.argmax(var.cumsum() > k) + 1
        return P.T[:idx], var[:idx]
    else:
        raise Exception("K debe ser un número entre 0 y 1, o un entero entre 1 y el número de columnas de A.")
```

```python
for k in [2,3,0.95,0.98]:
    C, var_exp = pca_eig(A, k)
    print(f"PCA, k = {k}:--------------------")
    print(C)
    print("Var Exp:--------------------------")
    print(var_exp)
```

```python
from sklearn.decomposition import PCA
for k in [2,3,0.95,0.98]:
    pca = PCA(n_components = k)
    pca.fit(A)
    C, var_exp = pca.components_, pca.explained_variance_ratio_
    print("PCA Sklearn: Iris, k = ", k)
    print("C: -----------------------------------------------------------")
    print(C)
    print("Var Exp: -----------------------------------------------------")
    print(var_exp)
    print("")

```
